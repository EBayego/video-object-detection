```python
import time
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
from googletrans import Translator
```

```python
def descargar_imagen(url, nombre_archivo):
    response = requests.get(url)
    if response.status_code == 200:
        with open(nombre_archivo, 'wb') as f:
            f.write(response.content)
            print(f"Imagen descargada: {nombre_archivo}")
    else:
        print(f"No se pudo descargar la imagen desde {url}")
```

```python
def descargar_imagenes_objeto(objeto, cantidad=100):
    url = f"https://www.google.com/search?q={objeto}&tbm=isch"

    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36",
    "Connection": "keep-alive",
    }
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        # Crear una carpeta con el nombre del objeto
        carpeta_objeto = objeto.replace(" ", "_")
        os.makedirs(carpeta_objeto, exist_ok=True)

        soup = BeautifulSoup(response.text, 'html.parser')
        imagenes = soup.find_all("img", limit=cantidad)
        print(imagenes)
        for i, imagen in enumerate(imagenes, 1):
            imagen_url = imagen['src']
            # Verificar si la URL tiene un esquema
            if not imagen_url.startswith('http'):
                imagen_url = urljoin(url, imagen_url)
            nombre_archivo = f"{carpeta_objeto}/{objeto}_{i}.jpg"
            descargar_imagen(imagen_url, nombre_archivo)
    else:
        print("Error al conectarse a Google Images")
```

```python
def descargar_imagenes_objeto_idiomas(objeto, cantidad=100):
    i = 0
    idiomas = ["es", "en", "ca", "de", "it", "nl", "ru", "zh-CN", "ja"]
    translator = Translator()
    for idioma in idiomas:
        print(idioma)
        translated_word = translator.translate(objeto, dest=idioma)
        print(translated_word.text)
        url = f"https://www.google.com/search?q={translated_word.text.replace(' ', '+')}&tbm=isch"
    
        headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36",
        "Connection": "keep-alive",
        }
        response = requests.get(url, headers=headers)
    
        if response.status_code == 200:
            carpeta_objeto = objeto.replace(" ", "_")
            os.makedirs(carpeta_objeto, exist_ok=True)
    
            soup = BeautifulSoup(response.text, 'html.parser')
            imagenes = soup.find_all("img", limit=cantidad)
            primera_imagen_google = True
            for imagen in imagenes:
                if primera_imagen_google:
                    primera_imagen_google = False
                    continue
                imagen_url = imagen['src']
                if not imagen_url.startswith('http'):
                    imagen_url = urljoin(url, imagen_url)
                nombre_archivo = f"{carpeta_objeto}/{carpeta_objeto}_{i}.jpg"
                i += 1
                descargar_imagen(imagen_url, nombre_archivo)
        else:
            print("Error al conectarse a Google Images")
```
